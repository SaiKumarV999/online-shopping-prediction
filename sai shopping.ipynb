```python
x=new_data.drop(["Revenue"],axis=1)

y=new_data["Revenue"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
teja=LinearRegression()
teja.fit(x_train,y_train)
```

```python

y_pred=teja.predict(x_test)


```

```python
print("training accuracy:",teja.score(x_train,y_train))
print("testing accuracy:",teja.score(x_test,y_test))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print("r2eror:",r2_score(y_pred,y_test))
print("MAE:",mean_absolute_error(y_pred,y_test))
print("MSE:",mean_squared_error(y_pred,y_test))
                    
```

```python
y_test=list(y_test)
```

```python
plt.figure(figsize=(20,20))
plt.plot(y_test,label="test",color="blue")
plt.plot(y_pred,label="pred",color="green")
plt.title("Revenue")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
tej=RandomForestRegressor()
```

```python
tej.fit(x_train,y_train)
y_pred=tej.predict(x_test)
```

```python
print("training accuracy:",tej.score(x_train,y_train))
print("testing accuracy:",tej.score(x_test,y_test))
```

```python
plt.figure(figsize=(20,8))
plt.plot(y_test,label="test",color="blue")
plt.plot(y_pred,label="pred",color="green")
plt.title("Revenue")
plt.legend()
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
kumar=GaussianNB()
```

```python
kumar.fit(x_train,y_train)
y_pred=kumar.predict(x_test)
```

```python
print("training accuracy:",kumar.score(x_train,y_train))
print("testing accuracy:",kumar.score(x_test,y_test))
```

```python
plt.figure(figsize=(20,20))
plt.plot(y_test,label="test",color="blue")
plt.plot(y_pred,label="pred",color="green")
plt.title("Revenue")
plt.legend()
plt.show()
```
